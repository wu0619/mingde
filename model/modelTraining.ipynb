```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
handsUp = pd.read_csv("handsUp.csv", index_col=0)
headLeft = pd.read_csv("headLeft.csv", index_col=0)
headRight = pd.read_csv("headRight.csv", index_col=0)
```

```python
df = pd.concat([handsUp, headLeft, headRight])
```

```python
df
```

```python
X = df.iloc[:, :99].to_numpy()
y = df.iloc[:, 99:].to_numpy().flatten()
print(X.shape)
print(y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
from joblib import dump
dump(svc, 'model.joblib')
```
