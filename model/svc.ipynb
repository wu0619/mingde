```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
data_path = "./data/"
```

```python
pose_00 = pd.read_csv(data_path + "pose_00.csv", index_col=0)
pose_01 = pd.read_csv(data_path + "pose_01.csv", index_col=0)
pose_02 = pd.read_csv(data_path + "pose_02.csv", index_col=0)
pose_03 = pd.read_csv(data_path + "pose_03.csv", index_col=0)
pose_04 = pd.read_csv(data_path + "pose_04.csv", index_col=0)
pose_05 = pd.read_csv(data_path + "pose_05.csv", index_col=0)
pose_06 = pd.read_csv(data_path + "pose_06.csv", index_col=0)
```

```python
df = pd.concat([
    pose_00,
    pose_01,
    pose_02,
    pose_03,
    pose_04,
    pose_05,
    pose_06,
], ignore_index=True)
```

```python
df
```

```python
df.describe()
```

```python
X = df.iloc[:, :99].to_numpy()
y = df.iloc[:, 99:].to_numpy().flatten()
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
svc = SVC(kernel='linear', probability=True)
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```
